```python
import torch
import cv2
from ultralytics import YOLO
```

```python
model =  YOLO('sample_data/yolov8n.pt')
# model.summary()
```

```python
classes = model.names
```

```python
# cap = cv2.VideoCapture(0)  #camera access

cap = cv2.VideoCapture('/content/drive/MyDrive/AI/Deep Learning/video.mp4')

fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter('/content/drive/MyDrive/AI/Deep Learning/output_video.mp4',fourcc,24,(1280,720))

while cap.isOpened():
  success, img = cap.read()
  if not success:
    cap.release()
    cv2.destroyAllWindows()
    break
  width = img.shape[1]
  height = img.shape[0]

  results = model(img)
  for result in results:
    objects = result.boxes.cls
    bboxes = result.boxes.xyxyn
    confidences = result.boxes.conf
    for obj, box,conf in zip(objects,bboxes,confidences):
      conf = conf.tolist()
      if conf < 0.3:
        continue
      conf = int(conf * 100)
      x1,y1 = int(box[0] * width), int(box[1]*height)
      x2,y2 = int(box[2] * width), int(box[3]*height)
      label = classes[obj.tolist()]
      cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
      text = f'{label} ({conf})%'
      cv2.putText(img,text,(x1,y1+30),cv2.FONT_HERSHEY_PLAIN,1,(0,255,0),1) # 1 =  font size , color, thickness

  out.write(img)
  # cv2.imshow('image',img)

  if cv2.waitKey(1) & 0xff == 27:
    cap.release()
    cv2.destroyAllWindows()
    break



```
